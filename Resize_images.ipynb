```python
from PIL import Image
import os, sys
```

```python
def resizeImage(infile, output_dir="", size=(128,128)):

    os.chdir("absolute path to the folder containing output dirs"+output_dir)
    
    try :
        #print("0")
        im = Image.open(infile)
        print(infile)
        im2=im.resize(size, Image.ANTIALIAS)
        #print("2  ",im2.size)
        os.chdir("./"+output_dir++"/"+output_dir+"_resized")
        im2.save(output_dir+infile)
        #print("3")
    except IOError:
        print("cannot reduce image for ", infile)

```

```python
#all the input dir sould be in one folder and output dirs in another folder

dir1 = ["Basal_cell_cancer",
        "melanoma",
        "Squamous_cell_carcinoma",
        "Solar_Lentigo",
        "Seborrhoeic_keratosis"
        ]
#example of input dir is provided above

output_dirs=['Basal_cell_cancer_resized',
             'melanoma_resized',
             'Squamous_cell_carcinoma_resized',
             'Solar_Lentigo_resized',
             'Seborrhoeic_keratosis_resized'
            ]
#example of output dir is given abouve


os.chdir("absolute path to the folder containing all the input dirs")


for output_dir in output_dirs:
    os.mkdir(output_dir)

#os.mkdir(output_dir)
for i in range(len(dir1)):
    for file in os.listdir(dir1[i]):
        #print(file)
        resizeImage(file,output_dir=dir1[i])
    
    print("-----------------------------------------------------------")
    os.chdir("absolute path to the folder containing all the input dirs")




    
    
# for file in os.listdir(dir1):
#     resizeImage(file,output_dir=output_dir)
```
